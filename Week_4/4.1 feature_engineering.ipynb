```python
data = pd.read_csv('labeled_data.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
# we should look manually what's inside every column
data['data_dump'].describe()
```

```python
# drop irrelated columns
data = data.drop(columns = ['Unnamed: 0', 'sysTime', 'GPSTime', 'Long', 'ElapsedRealtimeNanos', 'HasSpeedAccuracy', 'SpeedAccuracy', 'HasBearingAccuracy', 'data_dump'])
```

```python
data.head()
```

```python
# deal with BearingAccuracy's nan value

from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)
imputer.fit(np.array(data['BearingAccuracy'])[:,np.newaxis])
```

```python
data['BearingAccuracy'] = imputer.transform(np.array(data['BearingAccuracy'])[:,np.newaxis])
```

```python
data.head()
```

```python
# deal with Provider

from sklearn import preprocessing
encoder =preprocessing.LabelEncoder()
```

```python
encoder.fit(np.array(data['Provider']))
data['Provider'] = encoder.transform(np.array(data['Provider']))
```

```python
data.head(5)
```

```python
data = data.drop(columns = ['id'])
```

```python
for i in range(data.shape[1] - 1):
    data.iloc[:,i] = (data.iloc[:,i] - data.iloc[:,i].mean()) / data.iloc[:,i].std()
```

```python
data.to_csv('transformed_data.csv')
```
