```python
data = pd.read_csv('local_time_attack.csv')
```

```python
data['Start_Hour'] = data['Start'].apply(lambda x: int(str(x)[:-2]))
data['Start_Minute'] = data['Start'].apply(lambda x: int(str(x)[-2:]))

data['End_Hour'] = data['End'].apply(lambda x: int(str(x)[:-2]))
data['End_Minute'] = data['End'].apply(lambda x: int(str(x)[-2:]))
```

```python
data
```

```python
data_one = data.iloc[0]
data_end = data.iloc[-1]
```

```python
single_min = (data_one['Epoch End'] - data_one['Epoch Start'])/((data_one['End_Hour'] * 60 + data_one['End_Minute']) - (data_one['Start_Hour'] * 60 + data_one['Start_Minute']))
```

```python
single_min
```

```python
start_epoch = data_one['Epoch Start'] - single_min * (data_one['End_Hour'] * 60 + data_one['End_Minute'])
end_epoch = start_epoch + single_min * 72 * 60
```
